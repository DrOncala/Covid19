```python
import pandas as pd
from numpy import *
import torch
import matplotlib.pyplot as plt
import geopandas as gpd
```

```python
url_confirmed = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv"
url_deths = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv"
url_recovered = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv"
url_countries = "https://raw.githubusercontent.com/AnthonyEbert/COVID-19_ISO-3166/master/JohnsHopkins-to-A3.csv"
data_confirmed = pd.read_csv(url_confirmed)
data_deths = pd.read_csv(url_deths)
data_recovered = pd.read_csv(url_recovered)
data_countries = pd.read_csv(url_countries)

days=data_confirmed.shape[1]-4

print("matrix shape:" ,data_confirmed.shape)
#data_confirmed[data_confirmed['Country/Region']=='China']
#print(data_confirmed.iloc[:,0:2].values)
data_confirmed.head(5)
```

```python
from numpy import *
from scipy.integrate import odeint
import matplotlib.pyplot as plt


tao0 = 2            #characteristic time of the pandemic growth
Nt0= 10000           #total number of people who could be infected
t = linspace(0,50,100)  #time points
alpha=0.01

def Nt(t, Nt0=Nt0):
    Nt=Nt0-N_an(t)
    return Nt
def tao(t):
    tao= tao0+alpha*t
    return tao


##### function that returns dN/dt eq.(1) ########
def dNdt(N,t):   #N: total number of infections
    dNdt = (1-N/Nt0)*N/tao0
    return dNdt
# -> initial condition
N0 = 1
# -> solve ODE
N = odeint(dNdt,N0,t)
##################################################

##### function that returns df/dt eq.(3) ########
def dfdt(f,t):   
    tao0 = 2
    dfdt = f*(1-f)/tao0
    return dfdt
# -> initial condition
C1=10
f0 = (1+exp(C1))**(-1)
# -> solve ODE
f = odeint(dfdt,f0,t)
################################################

##### function that returns f_an eq.(6) #######
def f_an(x,f0=f0): 
    f_an = f0*exp(x)/(f0*exp(x)+1)
    return f_an
##################################################

##### function that returns N_an eq.(7) ########
def N_an(x,f0=f0,Nt=Nt0):
    N_an = Nt*f0*exp(x)/(f0*exp(x)+1)
    return N_an
##################################################

x=t/tao0
```

```python
fig = plt.subplots()
plt.plot(t, N, '-b', label='Nuberical')
plt.plot(t, N_an(x), '--r', label='Analytical')
#plt.axis('equal')
leg = plt.legend()
plt.xlabel('day')
plt.ylabel('N(t)')
plt.title("plot1")
print("using random constants")
```

```python
fig = plt.subplots()
plt.plot(t, f, '-b', label='Nuberical')
plt.plot(t, f_an(x), '--r', label='Analytical')
#plt.axis('equal')
leg = plt.legend()
plt.xlabel('day')
plt.ylabel('f(t)')
plt.title("plot2")
print("using random constants")
```

```python

def DeltaN(t):  
    DeltaN = (Nt0*f0*exp(t/tao0))/(tao0*(1+f0*exp(t/tao0)**2))
    return DeltaN

def DeltaNi(t):  
    DeltaNi = (Nt(t)*f0*exp(t/tao(t)))/(tao(t)*(1+f0*exp(t/tao(t)))**2)
    return DeltaNi
    
```

```python
fig = plt.subplots()
plt.plot(t, DeltaN(t), '-b', label='DeltaN')
plt.plot(t, DeltaNi(t), '--r', label='DeltaNi')
leg = plt.legend()
plt.xlabel('day')
plt.ylabel('DeltaN(t)')
plt.title("plot3")
print("using random constants")
```
